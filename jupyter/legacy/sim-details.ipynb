```python
#mass is in units of 10e10 M_solar
#distances are in kpc
from decimal import Decimal
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import h5py
```

```python
file = h5py.File('/home/ryan/Data/Data2/run_CDM_L3N512/snap_002.hdf5', 'r')
```

```python
def dump_info(name, obj):
    print ("{0} :".format(name))
    try:
        print ("   .value: {0}".format(obj.value))

        for key in obj.attrs.keys():
            print ("     .attrs[{0}]:  {1}".format(key, obj.attrs[key]))
    except:
        pass
```

```python
file.visititems(dump_info)
```

```python
mass_res_in_msol = file["Header"].attrs['MassTable'][1] * 10e10
'%.2E' % Decimal(mass_res_in_msol)
```
