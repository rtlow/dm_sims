```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import json

import sys
sys.path.append('../../py/')

from cosmoSim import cosmoSim
```

```python
matplotlib.rc('xtick', labelsize=24)
matplotlib.rc('ytick', labelsize=24)
matplotlib.rcParams['font.size']=24

ALPHA=0.6
FIGSIZE=[12,12]
DPI=300

redshift = 1
```

```python
outpath = '../../plots/plots_L3N256_densities/'

try:
    os.mkdir(outpath)
except:
    print(f'{outpath} already exists!')


runs = [
    'run_CDM_L3N256_DM',
    'run_CDM_L3N256_HY',
    'run_2cDM_L3N256_DM_power00_sigma1',
    'run_2cDM_L3N256_HY_power00_sigma1',
    'run_2cDM_L3N256_DM_powerm2m2_sigma1',
    'run_2cDM_L3N256_HY_powerm2m2_sigma1'
    ]
colors = [
    'r',
    'b',
    'g',
    'tab:purple'
]

CDM_runs = [
    'run_CDM_L3N256_DM',
    'run_CDM_L3N256_HY'
    ]
tcDM_runs = [
    'run_2cDM_L3N256_DM_power00_sigma1',
    'run_2cDM_L3N256_HY_power00_sigma1',
    'run_2cDM_L3N256_DM_powerm2m2_sigma1',
    'run_2cDM_L3N256_HY_powerm2m2_sigma1'
]
```

```python
fig, ax = plt.subplots(1,1, figsize=FIGSIZE)


for run_n in runs:

    run = cosmoSim(run_n)
    try:
        bins, density = run.load_mass_density(redshift, 0, partType='all')
    except:
        continue

    ax.plot(bins, density, label=run.plot_label)
ax.legend()
ax.set_ylabel("$\\rho\left(r\\right) \, (M_{\odot}/kpc^{3})$")
ax.set_xlabel("$r \, (kpc)$")
ax.set_yscale("log")
ax.set_xscale("log")
ax.grid(True, which="both", ls="-")
```
