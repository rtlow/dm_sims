```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc('xtick', labelsize=24)
matplotlib.rc('ytick', labelsize=24)
matplotlib.rcParams['font.size']=24
```

```python
sigmas = [ x for x in np.geomspace(0.5, 5, num=10)]
sigmas.append(1)
sigmas.append(10)
sigmas.append(0.1)

c_norm = matplotlib.colors.LogNorm(vmin=np.amin(sigmas), vmax=np.amax(sigmas))
c_map = matplotlib.cm.viridis

s_map = matplotlib.cm.ScalarMappable(cmap=c_map, norm=c_norm)
```

```python
m2m2 = lambda v, sigma: sigma * (v/100)**(-2)

zz = lambda v, sigma: sigma * (v/100)**(-1)
```

```python
v = np.geomspace(1,1000, num=10000)
```

```python
fig, ax = plt.subplots(1,1, figsize=[12,12])

for s in sigmas:
    ax.plot( v, zz(v, s), color=s_map.to_rgba(s))
    ax.plot( v, m2m2(v, s), color=s_map.to_rgba(s), linestyle='dashdot')

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("$V$ (km/s)")
ax.set_ylabel("$\sigma (v)$ (cm$^2$ g$^{-1}$)")
ax.plot([],[], label='[0, 0]', color='k')
ax.plot([],[], label='[-2, -2]', color='k', linestyle='dashdot')
ax.grid(True, which="both", ls="-")
ax.legend()
fig.colorbar(s_map, label='$\sigma_{0}$ [cm$^2$ g$^{-1}$]', ax=ax)
plt.show()
```
