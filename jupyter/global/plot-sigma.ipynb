```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc('xtick', labelsize=24)
matplotlib.rc('ytick', labelsize=24)
matplotlib.rcParams['font.size']=24
```

```python
v = np.geomspace(1,1000, num=10000)

sm2 = 1 * (v/100)**(-2)
s0 = 1 * (v/100)**(-1)
```

```python
fig, ax = plt.subplots(1,1, figsize=[12,12])

ax.plot(v, s0, label='(0, 0)')
ax.plot(v, sm2, label='(-2,-2)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("$V$ (km/s)")
ax.set_ylabel("$\sigma_{0}$ (g/cm^3)")
ax.grid(True, which="both", ls="-")
ax.legend()
```
